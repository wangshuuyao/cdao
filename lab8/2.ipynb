```python
from pylab import *

# 我们要生成的随机点的数量。
n = 1024

# 生成n个随机数，服从平均值为0，标准差为1的正态分布，赋值给X。
X = np.random.normal(0, 1, n)

Y = np.random.normal(0, 1, n)

# 使用scatter函数创建一个散点图，其中X轴坐标为X数组中的值，Y轴坐标为Y数组中的值。
scatter(X, Y)

show()
```

```python
from pylab import *
n=12 #12个柱状数据
X=np.arange(n)#生成0到n-1的数组，表示x轴
Y=(1-X/float(n)*np.random.uniform(0.5,1.0,n))#生成Y数据，在0.5-1.0均匀分布
bar(X,+Y,facecolor='pink',edgecolor='blue')#绘制柱状图
for x,y in zip(X,Y): # 通过循环，将每个 Y的值显示在对应的柱子上方。  
    text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')
ylim(0,+1.25)#y的范围
show()
```
